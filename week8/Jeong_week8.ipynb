```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../week7/creditcard.csv')
df.head()
```

```python
df.describe()
```

```python
#normalizing time, amount
time = df['Time']; amount = df['Amount']
df['Time'] = (time - time.mean()) / time.std()
df['Amount'] = (amount - amount.mean()) / amount.std()
```

```python
df['Class'].value_counts().plot.bar()
```

```python
fig,ax = plt.subplots(6,5,figsize=(40,48))
columns = df.columns
for i in range(6):
    for j in range(5):
        data = df[columns[i * 5 + j]]
        ax[i][j].hist(data, bins=100, density=True, alpha=0.6, color='b')
        ax[i][j].set_title(columns[i * 5 + j])
```

```python
from scipy.stats import norm
norm_dist = norm(loc = df['V18'].mean(), scale = df['V18'].std())
x = df['V18']
y = norm_dist.pdf(x)
```

```python
prob_df = pd.DataFrame()
for c in df.columns[1:-1]: #시간과 라벨의 확률을 구할 필요는 없다.
    rv = norm(loc = df[c].mean(), scale = df[c].std())
    prob_df[c] = rv.pdf(df[c])
prob_df
```

```python
none_fraud = prob_df.loc[df['Class'] == 0]
train_size = int(len(none_fraud))
train_X = none_fraud.iloc[:train_size].sample(frac=1)
test_X = none_fraud.iloc[train_size:].sample(frac=1)
fraud = prob_df.loc[df['Class'] == 1]
```

```python
train_X = (train_X - train_X.mean()) / train_X.std()
fraud = (fraud - fraud.mean()) / fraud.std()
```

```python
none_fraud_prob = train_X.cumprod(axis=1).iloc[::,-1]
none_fraud_prob
```

```python
plt.scatter(none_fraud_prob.index, none_fraud_prob.values)
plt.ylim(-500,500,20)
```

```python
fraud_prob = fraud.cumprod(axis=1).iloc[::,-1]
fraud_prob
```

```python
plt.scatter(fraud_prob.index, fraud_prob.values)
plt.ylim(-500,500,20)
```
